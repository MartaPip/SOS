```python
import pandas as pd
import time
from datetime import datetime

# Create a sample DataFrame with an 'ArrivalTime' column
data = {'Name': ['Job1', 'Job2', 'Job3'],
        'ArrivalTime': ['2023-11-22 10:00:00', '2023-11-22 11:30:00', '2023-11-22 12:45:00'],
        'car':[2,4,6]}

df = pd.DataFrame(data)

# Convert 'ArrivalTime' column to datetime
df['ArrivalTime'] = pd.to_datetime(df['ArrivalTime'])

# Print the DataFrame
print("DataFrame:")
print(df)

# Function to check and print when the time is reached
def check_arrival_time():
    for _, row in df.iterrows():
        current_time = datetime.now()
        arrival_time = row['ArrivalTime']

        if current_time >= arrival_time:
            print(f"One job arrived at {current_time} for {row['Name']}")

# Check for arrival time every 1 minute
while True:
    check_arrival_time()
    time.sleep(30)  # sleep for 1 minute
```
